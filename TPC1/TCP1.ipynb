2. Create a function that given a string 's', returns how many 'a' and 'A' characters are present in it.

```python
def conta_a(s):
    contagem=0
    for letra in s:
        if letra.lower()=='a':
            contagem+=1
    return contagem

print(conta_a('exercicio TPC1'))
print(conta_a('exercicio da aula 1'))
print(conta_a('exercicio da Aula 1'))
```

```python
def conta_aA(s):
    contagem={'a': 0, 'A': 0}
    for letra in s:
        if letra=='a':
            contagem["a"]+=1
        if letra=='A':
            contagem["A"]+=1
    return contagem

print(conta_aA('exercicio TPC1'))
print(conta_aA('exercicio da aula 1'))
print(conta_aA('exercicio da Aula 1'))
```

3. Create a function that given a string 's', returns the number of vowels there are present in it.

```python
def vogais(s):
    vogais=['a', 'e', 'i', 'o', 'u']
    contagem=0
    for letra in s:
        if letra.lower() in vogais:
            contagem+=1

    return contagem

print(vogais('exercicio TPC1'))
```

6. Verifica se uma string é capicua.

```python
def capicua(s):
    texto=s.lower()
    acentos = {
        'á':'a','à':'a','ã':'a','â':'a','ä':'a',
        'é':'e','è':'e','ê':'e','ë':'e',
        'í':'i','ì':'i','î':'i','ï':'i',
        'ó':'o','ò':'o','õ':'o','ô':'o','ö':'o',
        'ú':'u','ù':'u','û':'u','ü':'u',
        'ç':'c'
    }

    for a, b in acentos.items():
        texto = texto.replace(a, b)

    pontuacao = '.,;:!?()[]{}"\'-_/\\|@#$%&*+= '
    for p in pontuacao:
        texto = texto.replace(p, '')

    if texto==texto[::-1]:
        print('É capicua')
    else:
        print('Não é capicua')
    return

capicua('123454321')
capicua('exercicio TPC1')
capicua('exercicio TPC Aula 1 1aluaCPT oicicrexe')
capicua("Socorram-me subi no ônibus em Marrocos")
```

7. Verifica se duas strings estão balanceadas (Duas strings, s1 e s2, estão balanceadas se todos os caracteres de s1 estão presentes em s2).

```python
def balanceadas(s1, s2):
    for letra in s1:
        if letra not in s2:
            return False
        else:
            return True

print(balanceadas('exercicio TPC1', 'exercicio'))
print(balanceadas('exercicio TPC1', 'Aula'))
```

8. Calcula o número de ocorrências de s1 em s2.

```python
def ocorrencias_s1(s1, s2):
    contagem=0
    for i in range(len(s2)-len(s1)+1):
        if s2[i:i+len(s1)].lower()==s1.lower():
            contagem+=1
    return contagem

print(ocorrencias_s1('exercicio', 'exercicio TPC1'))
print(ocorrencias_s1('Aula', 'Aulaulaoaula'))
```

9. Verifica se s1 é anagrama de s2.

- "listen" e "silent": deve imprimir true
- "hello" e "world": deve imprimir false

```python
def anagrama(s1,s2):
    if sorted(s1.lower())==sorted(s2.lower()):
        return True
    else:
        return False

print(anagrama("listen", 'silent'))
print(anagrama("hello", "world"))
```
